<font color='Green'> پرسش: اگر تعداد خوشه‌ها را برای داده‌های قبل 2 در نظر بگیریم، نمودار قبلی را برای این حالت رسم کنید. دقت کنید که <code>random_state=42</code> باشد. </font>

<div dir=rtl>
<h2>روش آرنج (elbow)</h2>
شما در درس با روش آرنج آشنا شده‌اید.

<font color='Green'> پرسش: روش آرنج را برای مقدار خوشه‌های 1 تا 10 حساب کنید و نمودار میزان تابع هزینه را برای هر خوشه رسم کنید. برای هر مرحله <code>random_state</code> را برابر با صفر در نظر بگیرید.  
</div>

```python
# your code here
elbow_lf = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(X)
  elbow_lf.append(kmeans.inertia_)

plt.plot(range(1, 11), elbow_lf, c='g')
plt.xticks(range(1, 11))
plt.xlabel("# of clusters")
plt.show()
```

<h2>خوشه‌بندی‌های پیچیده‌تر</h2>
همیشه خوشه‌بندی‌ها به این صورت نیست. همچنین k-means تنها الگوریتم خوشه‌بندی نیست. در زیر توزیع‌های مختلف داده‌های مختلف با روش k-means رسم شده‌است.

 شما می‌توانید مشاهده کنید که در کدام حالت این الگوریتم خوب عمل می‌کند.

تمامی این‌ کدها از سند مطالب <code>sklearn</code> استفاده شده است. شما می‌توانید این اسناد را در <a href="https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_assumptions.html"> اینجا</a> مشاهده کنید.


همچنین می‌توانید الگوریتم‌های دیگر را نیز در اسناد <code>sklearn</code> مشاهده کنید.
</div>

```python
sns.set_style('white')

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# Incorrect number of clusters
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")

# Anisotropicly distributed data
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.subplot(222)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.title("Anisotropicly Distributed Blobs")

# Different variance
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)

plt.subplot(223)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
plt.title("Unequal Variance")

# Unevenly sized blobs
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_filtered)

plt.subplot(224)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Unevenly Sized Blobs")

plt.show()
```

<font color='Green'> پرسش: علت استفاده از KNN برای پیدا کردن اپسیلون بهینه را شرح دهید. سپس با رسم گراف k-distance  و با پیروی از الگوی آرنج مقدار اپسیلون بهینه را   بدست آورید.

<font color='Green'>  برای رسم گراف k-distace ‌می‌توانید از ماژول NearestNeighbors در کنابخانه Scikit-learn و برای مشخص کردن مقدار اپسیلون از کتابخانه kneed استفاده کنید.

<font color='white'>  جواب:
به این دلیل از الگوریتم KNN استفاده می‌کنیم تا بتوانیم یک ایده‌ی  مناسب از میانگین فاصله‌ی هر نقطه با K نقطه‌ی نزدیکش داشته باشیم تا در نتیجه مقدار eps را هوشمندانه انتخاب کنیم.
</font>


</div>




```python
!pip install --upgrade kneed

from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import v_measure_score
from kneed import KneeLocator
import pandas as pd 
```

```python
def DBSCAN_Kneepoint(dataset):

  # Calculate the average distance between each point in the data set and its 20 nearest neighbors
  neighbors = NearestNeighbors(n_neighbors=10)
  neighbors_fit = neighbors.fit(dataset)
  distances, indices = neighbors_fit.kneighbors(dataset)

  # Sort distance values by ascending value and plot
  distances = np.sort(np.average(distances, axis=1))
  # distances = distances[:,1]
  plt.title("k-distance graph")
  plt.plot(distances)

  i = np.arange(len(distances))
  kl = KneeLocator(i, distances, S=1, curve='convex', direction='increasing', interp_method='polynomial')
  fig = plt.figure()
  kl.plot_knee()
  plt.xlabel("Points")
  plt.ylabel("Distance")

  print("eps: ", distances[kl.knee])
  return distances[kl.knee]
```

```python
epsilon_ansio = DBSCAN_Kneepoint(X_aniso)
```

```python
epsilon_varied = DBSCAN_Kneepoint(X_varied)
```

<font color='Green'> پرسش: در این مرحله با در نظر گرفتن اپسیلون بدست آمده و با آزمودن مقادیر مختلف برای  پارامتر MinPts، بهترین حالت برای خوشه‌بندی را با استفاده از رنگ‌بندی مناسب رسم کنید. 

</div>




```python
# your code here
def MinPts_DBSCAN(eps, dataset):
  v_measure_array = []
  fig = plt.figure(figsize=(20, 10))
  fig.subplots_adjust(hspace=.5, wspace=.2)
  i = 1
  for j in range(3, 13):
      db = DBSCAN(eps=eps, min_samples=j).fit(dataset)
      core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
      core_samples_mask[db.core_sample_indices_] = True
      labels = db.labels_

      v_measure_array.append(v_measure_score(y, labels))
      
      ax = fig.add_subplot(2, 5, i)
      ax.text(1, 4, "minPts = {}".format(round(j)), fontsize=10, ha="center")
      sns.scatterplot(dataset[:,0], dataset[:,1], hue=["cluster-{}".format(x) for x in labels])
      
      i += 1

  print("v-measures: ", v_measure_array)
```

```python
MinPts_DBSCAN(epsilon_ansio, X_aniso)
```

```python
def DBSCAN_Algorithm(eps, minpts, dataset):
  db = DBSCAN(eps=eps, min_samples=minpts).fit(dataset)

  plt.scatter(dataset[:, 0], dataset[:, 1], c=db.labels_, label=y)

  labels = db.labels_
  num_clusters = len(set(labels))-(1 if -1 in labels else 0)

  print('number of clusters: ', num_clusters)
  print('v_measure: ', v_measure_score(y, labels))
```
